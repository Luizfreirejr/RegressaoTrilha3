```python
# Categorizando os países por continente
def categorize_continent(country):
    if country in ['Brazil', 'Argentina', 'Chile']:
        return 1
    elif country in ['USA', 'Canada']:
        return 20
    elif country in ['Germany', 'France']:
        return 40
    else:
        return 0

data['continent'] = data['country'].apply(categorize_continent)

data.head()
```

```python
def map_country_to_continent(country):
    if country in ["Brazil", "Argentina", "Chile"]:
        return 1
    elif country in ["USA", "Canada"]:
        return 20
    elif country in ["Germany", "France"]:
        return 40
    else:
        return 99

data['continent'] = data['country'].apply(map_country_to_continent)

```

```python
# Convertendo as strings com sufixos e evitar erro com valores não numéricos
def convert_to_numeric(value):
    if isinstance(value, str):
        # Lidar com sufixos como 'k' e 'M'
        if 'k' in value:
            try:
                return float(value.replace('k', '').strip()) * 1000
            except ValueError:
                return None  # Retorna None se não for um número válido
        elif 'M' in value:
            try:
                return float(value.replace('M', '').strip()) * 1000000
            except ValueError:
                return None
        # Tentando converter para float, retornando None se falhar
        try:
            return float(value)
        except ValueError:
            return None
    return value

# Aplicando a função em todas as colunas
data = data.applymap(convert_to_numeric)

# Calculando a correlação
correlation_matrix = data.corr()

# Exibe o heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd

# Convertendo os valores com sufixos 'k' e 'M' para números
def convert_to_numeric(value):
    # Verificando se o valor é uma string
    if isinstance(value, str):
        value = value.strip().lower()  # Remove espaços e converte para minúsculo
        if 'k' in value:  # Valores em milhar
            return float(value.replace('k', '').strip()) * 1000
        elif 'm' in value:  # Valores em milhão
            return float(value.replace('m', '').strip()) * 1000000
        elif value.replace('.', '', 1).isdigit():  # Verifica se é um número
            return float(value)
        else:
            return None  # Retorna None para valores não numéricos
    # Caso já for um número, retorna o valor
    elif isinstance(value, (int, float)):
        return value
    return None  # Outros tipos de dados não reconhecidos

data = pd.DataFrame({
    'col1': ['1.2k', '5k', 'abc', '2.5M'],
    'col2': ['10.5', '100k', '500', '2.5M'],
    'col3': ['100', '10M', 'yliejenner', '1k']
})

# Aplicando a função para converter valores para numérico
data = data.applymap(convert_to_numeric)

# Removendo linhas com valores faltantes
data = data.dropna()

# Exibindo o DataFrame resultante
print(data)
```
